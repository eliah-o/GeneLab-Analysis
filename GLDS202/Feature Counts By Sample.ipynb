```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
inputfile = open("../GLDS-202_Unnormalized_Counts.txt", "r")

header = True
firstLine = True
samples = []
sample_counts = []
for line in inputfile: 
    
    fields = line.strip().split("\t")
    fields = fields[1:]
    
    if header == True: 
        header = False
        
        for col in fields: 
            samples.append(col)
        
    else:
        
        if firstLine == True: 

            for col in fields: 
                sample_counts.append(float(col))
                
            firstLine = False
        else:
            for i in range(0, len(fields)): 
                sample_counts[i] += float(fields[i])
    
```

```python
y_pos = np.arange(len(samples))
 
plt.figure(num=None, figsize=(24, 8), dpi=80, facecolor='w', edgecolor='k')
plt.bar(y_pos, sample_counts, align='center')
plt.xticks(y_pos, samples, rotation=90)
plt.ylabel('Feature Count')
plt.title('Feature Counts per Sample - GLDS 202')

plt.show()
```
