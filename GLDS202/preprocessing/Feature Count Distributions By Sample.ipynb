# Plot Feature Count Distributions.
## Do any samples have distribution outliers?

```r
# Load Unnormalized Read Counts
unnormalized_counts = "../GLDS-202_Unnormalized_Counts.txt"
cts <- read.table(unnormalized_counts, sep="\t", row.names="GeneID", header=TRUE)
```

```r
# Set Figure Size
library(repr)
options(repr.plot.width=12, repr.plot.height=18)
```

```r
library(tidyr)
library(ggplot2)

# Plot the Log-transformed read counts of all samples.
ggplot(gather(log(cts)), aes(value)) + 
    geom_histogram(bins = 20) + 
    facet_wrap(~key, scales = 'free_x') + theme(aspect.ratio = 1) 
```
